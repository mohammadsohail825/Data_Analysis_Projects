   ### Step 3. Assign each to a variable called cars1 and cars2

```python
df1= pd.read_csv("Desktop/Cybrom/Python Classes/Pandas/Pandas-2/cars1 - cars1.csv")
df2 = pd.read_csv("Desktop/Cybrom/Python Classes/Pandas/Pandas-2/cars2 - cars2.csv")
```

### Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
df1 = df1.loc[:,"mpg":"car"] # loc funtion 3 perameter lega [is perameter par data lock hoga,loc start point:loc stop point] stop point ke bad ke columns nhi lega
df1.head()
```

### Step 5. What is the number of observations in each dataset?

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([df1,df2], ignore_index= True) # ignore_index=True se data vertical me na ho kar horzental (upar se niche) me add hoga, column similer hona chahiye
cars # join ke liye same type of data hona chahiye row and columns
```
